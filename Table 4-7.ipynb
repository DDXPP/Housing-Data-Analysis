```python
##Creating mapping for each sheet in excel file
LIHTC_data = pd.ExcelFile("2018-LIHTC-Tenant-Tables.xlsx")

sheets = {}
for sheet_name in LIHTC_data.sheet_names:
    sheets[sheet_name] = pd.read_excel("2018-LIHTC-Tenant-Tables.xlsx", sheet_name=sheet_name)
```

```python
##Table 4

##Dropping rows that are irrelevant 
sheets["Table 4"] = sheets["Table 4"].drop([0, 1, 58, 59, 60, 61, 62, 63, 64, 65, 66, 67, 68, 69, 70, 71, 72, 73, 74, 75, 76, 77, 78, 79, 80])

##Renaming cols to match actual dataset
sheets["Table 4"] = sheets["Table 4"].rename(columns={"Table 4: Race/Ethnicity of Heads of Household": "State"})
sheets["Table 4"] = sheets["Table 4"].rename(columns={"Unnamed: 1": "Properties Reported (%)"})
sheets["Table 4"] = sheets["Table 4"].rename(columns={"Unnamed: 2": "White Alone (%)"})
sheets["Table 4"] = sheets["Table 4"].rename(columns={"Unnamed: 3": "Black or African American Alone (%)"})
sheets["Table 4"] = sheets["Table 4"].rename(columns={"Unnamed: 4": "Asian Alone (%)"})
sheets["Table 4"] = sheets["Table 4"].rename(columns={"Unnamed: 5": "American Indian and Alaska Native Alone (%)"})
sheets["Table 4"] = sheets["Table 4"].rename(columns={"Unnamed: 6": "Native Hawaiian and Other Pacific Islander Alone (%)"})
sheets["Table 4"] = sheets["Table 4"].rename(columns={"Unnamed: 7": "Other (including Multiple Race) (%)"})
sheets["Table 4"] = sheets["Table 4"].rename(columns={"Unnamed: 8": "Hispanic (Any Race) (%)"})
sheets["Table 4"] = sheets["Table 4"].rename(columns={"Unnamed: 9": "Race or Ethnicity Not Reported (%)"})
sheets["Table 4"] = sheets["Table 4"].rename(columns={"Unnamed: 10": "Total (%)"})

##Getting rid of footnote
sheets["Table 4"] = sheets["Table 4"].replace(to_replace = "New York1", value = "New York")

##Dropping 'total' row
sheets["Table 4"] = sheets["Table 4"].drop([57])
sheets["Table 4"]
```

```python
##Table 7

##Dropping irrelevant rows
sheets["Table 7"] = sheets["Table 7"].drop([0, 1,58, 59, 60, 61, 62, 63, 64, 65])

##Renaming columns to match table
sheets["Table 7"] = sheets["Table 7"].drop(columns = ["Unnamed: 7", "Unnamed: 8", "Unnamed: 9"])
sheets["Table 7"] = sheets["Table 7"].rename(columns={"Table 7: Family Composition: Households with Children and Elderly Members": "State"})
sheets["Table 7"] = sheets["Table 7"].rename(columns={"Unnamed: 1": "All Household Members Reported (% of Households)"})
sheets["Table 7"] = sheets["Table 7"].rename(columns={"Unnamed: 2": "Head of Household (%)"})
sheets["Table 7"] = sheets["Table 7"].rename(columns={"Unnamed: 3": "All Reported Members (%)"})
sheets["Table 7"] = sheets["Table 7"].rename(columns={"Unnamed: 4": "At Least One Member < 18 (%)"})
sheets["Table 7"] = sheets["Table 7"].rename(columns={"Unnamed: 5": "At Least One Member >= 62 (%)"})
sheets["Table 7"] = sheets["Table 7"].rename(columns={"Unnamed: 6": "Reported Head of Household >= 62 (%)"})

##Getting rid of footnote
sheets["Table 7"] = sheets["Table 7"].replace(to_replace = "New York2", value = "New York")

##Replacing missing values w/ nan
sheets["Table 7"] = sheets["Table 7"].replace(to_replace = "Data Not Reported", value = np.nan) 

##Dropping 'total' row
sheets["Table 7"] = sheets["Table 7"].drop([57])

sheets["Table 7"]
```
